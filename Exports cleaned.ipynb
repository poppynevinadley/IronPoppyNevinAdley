```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

```python
#Read the exports of commodities to European Union file.
exp = pd.read_excel('EXP_2001_2019_20230306.xlsx')
exp = pd.DataFrame(exp)
```

```python
exp
```

```python
#Isolating the wood exports.
wood = exp["Descrição CUCI Subgrupo"].str.contains("madeira|carvão", case=False)
wood = exp[wood]
wood
```

```python
#renaming and deleting unecessary columns.
wood = wood.rename(columns={'Descrição CUCI Subgrupo':'Wood'})
wood = wood.drop(columns=['Bloco Econômico', 'Código CUCI Subgrupo'])
```

```python
#Function to rename the columns according to the years.
def year_func(df):
    for i, col in enumerate(df.columns):
        new_col = col.split()[0]
        df.rename(columns={col: new_col}, inplace=True)
    return df
```

```python
#Function to create a DataFrame for each commodity.
def product(df, name):
    product = df["Descrição CUCI Subgrupo"].str.contains(name, case=False)
    prod = df[product]
    prod = prod.drop(columns=['Bloco Econômico', 'Código CUCI Subgrupo'])
    prod = year_func(prod)
    prod = prod.sum().to_frame()
    prod = prod.drop(['Descrição'], axis=0)
    return prod
```

```python
#Apply the function for wood.
wood = product(exp, "madeira|carvão")
wood = wood.rename(columns={0:'Total wood exports'})
wood
```

```python
#Apply the function for coffee.
coffee = product(exp, 'Café')
coffee = coffee.rename(columns={0:'Total coffee exports'})
coffee
```

```python
#Apply the function for rubber.
rubber = product(exp, 'borracha')
rubber = rubber.rename(columns={0:'Total rubber exports'})
rubber
```

```python
##Apply the function for palm.
palm = product(exp, 'palma')
palm = palm.rename(columns={0:'Total palm exports'})
palm
```

```python
##Apply the function for soy.
soy = product(exp, 'soja')
soy = soy.rename(columns={0:'Total soy exports'})
soy
```

```python
##Apply the function for meat.
meat = product(exp, 'carne|bovino|couro|animais')
meat = meat.rename(columns={0:'Total meat exports'})
meat
```

```python
#File containing the cocoa exports number.
cocoa_df = pd.read_excel('EXP_cocoa.xlsx')
cocoa = product(cocoa_df, 'cacau')
cocoa = cocoa.rename(columns={0:'Total cocoa exports'})
cocoa
```

```python
#Concatenate all the commodities.
exports = pd.concat([meat, cocoa, wood, rubber, soy, coffee], axis=1)
```

```python
#Adjusting the index.
exports = exports.reset_index()
#create a file with the new DataFrame.
exports.to_csv('exports.csv')
exports = exports.rename(columns={'index':'year'})
exports.set_index('year', inplace=True)
exports = exports.iloc[::-1]
exports
```
