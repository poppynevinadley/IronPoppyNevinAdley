```python
#Pivot table, sum  and make the the city ID the column name.
df2 = df2.pivot_table(index='year', columns='geocode_ibge', values='areakm', aggfunc=sum).reset_index()
df2
```

```python
#Check if there are no cities with the same city ID.
df2.columns.unique()
```

```python
#the DataFrame shows the area of florest in each city per year. Subtracting one year by the previous we find the deforestation number.
df2.iloc[:, 1:] = df2.iloc[:, 1:].diff()
```

```python
##drop the first line. After the differencing it has no values.
df2 = df2.drop(df2.index[[0]])
df2
```

```python
#Set the years as index and sum to find the total deforestation.
df2_total = df2.set_index('year').transpose()
df2_total = df2_total.sum().to_frame()
```

```python
df2_total
```

```python
#rename to column.
df2_total = df2_total.rename(columns={0:'Deforestation Cerrado'})
```

```python
#Make the years column again.
df2_total = df2_total.reset_index()
df2_total
```

```python
#Save the new Data Frame as a csv file.
df2_total.to_csv('deforestation_cerrado_t.csv')
```

```python
#Preparing the Data Frame for finding the write names. Match the city name column to the one in the df_cities Data Frame.
df2 = df2.reset_index().rename(columns={'geocode_ibge': 'id_municipio'})
df2
```
